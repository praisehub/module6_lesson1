```python
#RIGHT MERGE BETWEEN ORDERS_DF AND CUSTOMERS_DF
right_merging = products_df.merge(customers_df, how = 'right')
right_merging
```

```python
#USING CONCATINATION METHOD OF MERGING TO JOIN THE THREE DATAFRAMES TOGETHER. THIS WILL INVOLVE DROPPING OF SOME COLUMNS
concatination = pd.concat([orders_df,products_df,customers_df])
concatination
```
